```python
import pandas as pd
import numpy as np

# Import the CSV file here and show the table without any changes
import pandas as pd
import numpy as np

taxiData = pd.read_csv("chicago_taxi_trips_2016_01.csv")
taxiData.head()
```

```python
column_mean = taxiData.mean()
```

```python
dataset2 = taxiData
dataset2 = dataset2.replace('NaN', np.nan)
dataset2 = dataset2.fillna(0)
dataset2.head()
```

```python
dataset2[['trip_date','trip_start_time']] = dataset2['trip_start_timestamp'].str.split(' ', expand=True)
dataset2.pop('trip_start_timestamp')
dataset2.head()
```

```python
dataset2[['date1','trip_end_time']] = dataset2['trip_end_timestamp'].str.split(' ', expand=True)
dataset2.pop('trip_end_timestamp')
dataset2.pop('date1')
dataset2.head()
```

```python
print("fare stastical data is\n",dataset2["fare"].describe())
print("mode ",dataset2["fare"].mode(),"\n")
print("tips stastical data is\n", dataset2["tips"].describe())
print("mode ",dataset2["tips"].mode(),"\n")
print("trip_minutes stastical data is\n", (dataset2["trip_seconds"].describe())/60)
print("mode ",dataset2["trip_seconds"].mode()/60,"\n")
print("trip_total stastical data is\n", dataset2["trip_total"].describe())
print("mode ",dataset2["trip_total"].mode(),"\n")
print("trip_miles stastical data is\n", dataset2["trip_miles"].describe())
print("mode ",dataset2["trip_miles"].mode(),"\n")
print(dataset2)
dataset2.describe()
```

```python
print(dataset2['pickup_census_tract'].value_counts())
print(dataset2['dropoff_census_tract'].value_counts())
print(dataset2['taxi_id'].value_counts())
print(dataset2['tolls'].value_counts())
print(dataset2['extras'].value_counts())
```

```python
dataset2.pop('pickup_census_tract')
dataset2.pop('tolls')
dataset2.pop('dropoff_census_tract')
dataset2.head()
```

```python
dataset2[['start_h','start_m','start_s']] = dataset2['trip_start_time'].str.split(':', expand=True)
dataset2.head()
```
